# Nested IF

```python
age = int(input("Enter the age: "))

if age >= 18 :
    citizen = input("Are you an Indian Citizen (yes/no): ")
    print(f"age is:{age}") 
    if citizen.lower() == "yes":
        print("Voter Id is granted")
    else:
        print(" To get voter ID Indian Citizen Ship is required")
else:
    print(f"Your age :{age} is under limit, you are not eligible to get voter id")
        
```

```python
age = int(input("Enter the age: "))
citizen = input("Are you an Indian Citizen (yes/no): ")
if age >= 18 :
    
    print(f"age is:{age}") 
    if citizen.lower() == "yes":
        print("Voter Id is granted")
    else:
        print(" To get voter ID Indian Citizen Ship is required")
else:
    print(f"Your age :{age} is under limit, you are not eligible to get voter id")
```

# match case (Switch case Java or C++)

```python
day = input("Enter the Day: ")
match day.lower():
    case "monday":
        print("start of the week")
    case "friday":
        print("weekend is coming")
    case "sunday":
        print("Holiday")
    case _:
        print("working day")
```
