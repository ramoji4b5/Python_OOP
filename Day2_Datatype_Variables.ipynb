# Custom / User-Defined Types: dataclasses


```python
from dataclasses import dataclass

@dataclass
class Product:
    id: int
    name: str
    price: float
    is_available: bool

# Instantiate a new custom data type
item1 = Product(101, "Wireless Mouse", 29.99, True)

print(item1.name)  # Outputs: Wireless Mouse
print(item1)       # Automatically generates a clean string representation
```

# Baking Applciaiton


```python
class BankAccountDetails:
    def __init__(self,name,balance,address,age,phoneno):
        self.name = name
        self.balance = balance
        self.address = address
        self.age = age
        self.phoneno = phoneno
        
    def display(self):
        print(f"Hello ! {self.name} your account balance is {self.balance}")
        print("Thanks you for visiting us")

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self,amount):
        self.balance -= amount
    
        
```

```python
ramoji_account = BankAccountDetails("Ramoji",1000,"Hyderabad",35,"8143121190")
```

```python
ramoji_account.display()
```

```python
ramoji_account.deposit(100)
```

```python
ramoji_account.display()
```

```python
ramoji_account.withdraw(500)
```

```python
ramoji_account.display()
```
